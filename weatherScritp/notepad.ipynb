Na początek kilka przykładów:

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt


# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(10000)

num_bins = 50
# the histogram of the data
n, bins, patches = plt.hist(x, num_bins, normed=1, facecolor='green', alpha=0.5)
# add a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

n_angles = 36
n_radii = 8

# An array of radii
# Does not include radius r=0, this is to eliminate duplicate points
radii = np.linspace(0.125, 1.0, n_radii)

# An array of angles
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)

# Repeat all angles for each radius
angles = np.repeat(angles[..., np.newaxis], n_radii, axis=1)

# Convert polar (radii, angles) coords to cartesian (x, y) coords
# (0, 0) is added here. There are no duplicate points in the (x, y) plane
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())

# Pringle surface
z = np.sin(-x*y)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0.2)

plt.show()
```

```python
import matplotlib.pyplot as plt
y_array = [1,2,3,5] # lista wartości na osi Y
plt.plot(y_array) # wartości Y, dla X od 0 (X0 = y_array[0], X1 = y_array[1], ...)
plt.xlabel(u'oś X')
plt.ylabel(u'oś Y')
plt.title(u'Przykładowy tytuł')
plt.grid(True) # siatka
plt.show() # rysuj
```

Dodatkowo w kolejnym przykładzie użyjemy funkcji axis(). Przekażemy jej jeden argument - listę z czteroma wartościam [xmin, xmax, ymin, ymax]

#Ćwiczenie 1

Narysuj wykres funkcji sinus na przedziale (-4, 4). Użyj funkcji numpy arange().

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(-4.0, 4.0, 0.01)
s = np.sin(t)
plt.plot(t, s)

plt.show()
```

W łatwy sposób możemy narysowac kilka wykresów na jednym rysunku.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
# na pierwszy rzut oka zapisy 'r--', 'bs', 'g^' mogą być niezrozumiałe:
# 'r--' -> red dashes
# 'bs' -> blue squares
# 'g^' -> green triangles

#Powyższy zapis możemy zapisać również w jednej lini:
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

#Legendy
Matplotlib pozwala na łatwe tworzenie legend do wykresów:

```python
import matplotlib.pyplot as plt

p1, = plt.plot([1,2,3])
l1 = plt.legend([p1], ["Legenda 1"], loc=2) # różne wartości argumentu loc zmieniają lokalizację legendy
plt.show()
```

Możemy również umieścić legendę poza samym wykresem i w różnych miejscach:

```python
import matplotlib.pyplot as plt


plt.plot([1,2,3], label="legenda 1")
plt.plot([3,2,1], label="legenda 2")
# Place a legend above this legend, expanding itself to
# fully use the given bounding box.
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2,  borderaxespad=0.)
plt.show()

plt.plot([1,2,3], label="legenda 3")
plt.plot([3,2,1], label="legenda 4")
# Place a legend to the right of this smaller figure.
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
import matplotlib.lines as mlines
import matplotlib.pyplot as plt

blue_line = mlines.Line2D([], [], color='blue', marker='*',
                          markersize=10, label='Blue stars')
red_squares = mlines.Line2D([], [], color='red', marker='s',
                          markersize=8, label='Red squares')
plt.plot(range(10), linestyle='-', marker='*',markersize=10, color='b')
plt.plot(range(10, 0, -1), linestyle='-', marker='s',markersize=8, color='r')

plt.legend(handles=[blue_line, red_squares])

plt.show()
```

### 1. Wyrażenia matematyczne
Matplotlib umożliwa nam w łatwy i czytelny sposób wyświetlanie wyrażeń matematycznych:

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t,s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)
plt.text(0.6, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',
         fontsize=20)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```

#3. Adnotacje

```python

import numpy.random
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(5,5))
fig.clf()

ax = fig.add_subplot(111)
ax.set_aspect(1)

x1 = -1 + numpy.random.randn(100)
y1 = -1 + numpy.random.randn(100)
x2 = 1. + numpy.random.randn(100)
y2 = 1. + numpy.random.randn(100)

ax.scatter(x1, y1, color="r")
ax.scatter(x2, y2, color="g")

bbox_props = dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9)
ax.text(-2, -2, "Sample A", ha="center", va="center", size=20,
        bbox=bbox_props)
ax.text(2, 2, "Sample B", ha="center", va="center", size=20,
        bbox=bbox_props)


bbox_props = dict(boxstyle="rarrow", fc=(0.8,0.9,0.9), ec="b", lw=2)
t = ax.text(0, 0, "Direction", ha="center", va="center", rotation=45,
            size=15,
            bbox=bbox_props)

bb = t.get_bbox_patch()
bb.set_boxstyle("rarrow", pad=0.6)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

plt.draw()
plt.show()
```
